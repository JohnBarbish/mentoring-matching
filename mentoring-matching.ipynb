### Now using our data to figure out the optimal compatibility yay!

```python
import pulp
```

```python
mentordf.columns
```

```python
mentors = mentordf['name'].to_list()
mentees = menteedf['name'].to_list()
max_mentees = mentordf['nmentees'].to_list()
mmd = dict(zip(mentors, max_mentees))
# a custom definition for research compatibility.
# can also translate it into a matrix for easier concatenation with other
# compatibility criteria
def compatibility(pair):
    (mentor_i, mentee_j) = pair
    mentorlist = mentordf.loc[mentordf['name'] == mentor_i, 'research'].values[0]
    menteelist = menteedf.loc[menteedf['name'] == mentee_j, 'research'].values[0]
    return len(set(mentorlist) & set(menteelist))/len(set(menteelist))
```

```python
# create lsit of all possible mentor-mentee pairs
possible_pairs = [tuple(c) for c in pulp.itertools.product(mentors, mentees)]
possible_pairs
```

```python
# create a binary decision variable to say whether or not a pair is in use or not
x0 = pulp.LpVariable.dicts(
    "pair", possible_pairs, lowBound=0, upBound=1, cat=pulp.LpInteger
)
```

```python
# set up problem
matching_model = pulp.LpProblem("Mentor-Mentee-Matching-Model", pulp.LpMaximize)
# want it to maximize the overall compatibility of all mentors and mentees
matching_model += pulp.lpSum([compatibility(pair)*x0[pair] for pair in possible_pairs])

# specify that each mentee can only be paired once
for mentee in mentees:
    matching_model += (
        pulp.lpSum(x0[pair] for pair in possible_pairs if mentee in pair) == 1, 
        "Mentee_%s_constraint" % mentee,
    )

# specify the maximum number of mentees for each mentor
for mentor in mentors:
    matching_model += (
        pulp.lpSum(x0[pair] for pair in possible_pairs if mentor in pair) <= mmd[mentor],
        "Mentor_%s_max_mentees" % mentor,
    )
    matching_model += (
        pulp.lpSum(x0[pair] for pair in possible_pairs if mentor in pair) >= 1,
        "Mentor_%s_min_mentees" % mentor,
    )
```

```python
matching_model
```

```python
matching_model.solve()
```

```python
pulp.value(matching_model.objective)
```

```python
print("The choosen pairs are out of a total of %s:" % len(possible_pairs))
for pair in possible_pairs:
    if x0[pair].value() == 1.0:
        print(pair)
```

# Now we work on improving the compatibility calculation by including additional details other than just research compatibility

```python
# add some preference data
mentordf.columns
```

```python
mentors = mentordf['name'].to_list()
mentees = menteedf['name'].to_list()
# set the maximum number of mentees per mentor to 2 since there aren't
# too many mentees
max_mentees = [min(i, 2.0) for i in mentordf['nmentees'].to_list()]
mmd = dict(zip(mentors, max_mentees))
# a custom definition for research compatibility.
# can also translate it into a matrix for easier concatenation with other
# compatibility criteria
def research_compatibility(pair):
    (mentor, mentee) = pair
    mentorlist = mentordf.loc[mentordf['name'] == mentor, 'research'].values[0]
    menteelist = menteedf.loc[menteedf['name'] == mentee, 'research'].values[0]
    return len(set(mentorlist) & set(menteelist))/len(set(menteelist))
```

```python
mentordf.columns
```

```python
for index, mentor in mentordf.iterrows():
    print(mentor['name'], mentor['preferences'])
```

```python
# defining a custom mentor compatibility function to allow preferences of mentors
def mentor_preferences(pair):
    (mentor, mentee) = pair
    mn = mentor
    if mn == 'mentor 10':
        # wants PhD students
        a = 0
        if 'PhD' == menteedf.loc[menteedf['name'] == mentee, 'degree'].values[0]:
            a += 5
        return a
    elif mn == 'mentor 5':
        # wants research compatibility
        return 2*compatibility(pair)
    elif mn == 'mentor 12':
        # wants PhD students
        a = 0
        if 'PhD' == menteedf.loc[menteedf['name'] == mentee, 'degree'].values[0]:
            a += 2
        if 'Thermal Sciences' in menteedf.loc[menteedf['name'] == mentee, 'research'].values[0]:
            a += 2
        return a
    else :
        return 1
```

```python
for index, mentee in menteedf.iterrows():
    print(mentee['name'], mentee['preferences'])
```

```python
# defining a custom mentee compatibility function to allow mentee compatibility
def mentee_preferences(pair):
    (mentor, mentee) = pair
    mn = mentee
    if mn == 'mentee 14':
        # wants research compatibility
        a = 0
        if 'Automotive and Transportation' in mentordf.loc[mentordf['name'] == mentor, 'research'].values[0]:
            a += 5
        return a
    elif mn == 'mentee 17':
        # wants international student
        a = 0
        if mentordf.loc[mentordf['name'] == mentor, 'international'].values[0]:
            a += 5
        return a
    elif mn == 'mentee 16':
        # wants PhD students
        a = 0
        if 'Thermal Sciences' in mentordf.loc[mentordf['name'] == mentor, 'research'].values[0]:
            a += 5
        return a
    else :
        return 1
```

```python
def overall_compatibility(pair):
    return research_compatibility(pair) + mentor_preferences(pair) + mentee_preferences(pair)
```

```python
# create list of all possible mentor-mentee pairs
possible_pairs = [tuple(c) for c in pulp.itertools.product(mentors, mentees)]
possible_pairs
```

```python
# create a binary decision variable to say whether or not a pair is in use or not
x = pulp.LpVariable.dicts(
    "pair", possible_pairs, lowBound=0, upBound=1, cat=pulp.LpInteger
)
# assign initial values to x based on previous run
for pair in possible_pairs:
    x[pair].setInitialValue(x0[pair].value())

# set up problem
matching_model = pulp.LpProblem("Mentor-Mentee-Matching-Model", pulp.LpMaximize)
# want it to maximize the overall compatibility of all mentors and mentees
matching_model += pulp.lpSum([overall_compatibility(pair)*x[pair] for pair in possible_pairs])

# specify that each mentee can only be paired once
for mentee in mentees:
    matching_model += (
        pulp.lpSum(x[pair] for pair in possible_pairs if mentee in pair) == 1, 
        "Mentee_%s_constraint" % mentee,
    )

# specify the maximum number of mentees for each mentor
for mentor in mentors:
    matching_model += (
        pulp.lpSum(x[pair] for pair in possible_pairs if mentor in pair) <= mmd[mentor],
        "Mentor_%s_max_mentees" % mentor,
    )
    matching_model += (
        pulp.lpSum(x[pair] for pair in possible_pairs if mentor in pair) >= 1,
        "Mentor_%s_min_mentees" % mentor,
    )

# add in that we don't want any mentor-mentee pairs in the same lab as a constraint
for mentor in mentors:
    for mentee in mentees:
        # get mentee advisor
        menteead = menteedf.loc[menteedf['name'] == mentee, 'advisor'].values[0]
        # get mentor advisor
        mentorad = mentordf.loc[mentordf['name'] == mentor, 'advisor'].values[0]
        # if they match, eliminate the pair
        if menteead == mentorad:
            pair = (mentor, mentee)
            x[pair].setInitialValue(0)
            x[pair].fixValue()
```

```python
# add good matches that we know we want
ptList = [('mentor 1', 'mentee 9'),
('mentor 8', 'mentee 1')]
for pt in ptList:
    x[pt].setInitialValue(1)
    x[pt].fixValue()
# add matches that we want to avoid
ptList = [('mentor 1', 'mentee 15')]
for pt in ptList:
    x[pt].setInitialValue(0)
    x[pt].fixValue()
```
